# Feature Engineering

## Convert the date str to yy-mm-dd

```python
import datetime

House_price_df['Date'] = pd.to_datetime(House_price_df['Date'], unit='D', origin=pd.Timestamp('1900-01-01'))
```

## Organizing data and viewing the first 10 rows

```python
House_price_df = House_price_df.sort_values('Date')
House_price_df.head(10)
```

```python
House_price_df['Year'] = pd.to_datetime(House_price_df['Date']).dt.year
House_price_df['Month'] = pd.to_datetime(House_price_df['Date']).dt.month
House_price_df['Week'] = pd.to_datetime(House_price_df['Date']).dt.isocalendar().week
House_price_df['day'] = pd.to_datetime(House_price_df['Date']).dt.day
```

```python
House_price_df = House_price_df.copy()
```

Let's review the dataset

```python
House_price_df
```

## Line graph of the mean price for each month

```python
House_price_df_monthly = House_price_df.groupby(pd.Grouper(key='Date', freq='M')).mean()
fig, ax = plt.subplots(figsize=(8, 5))
plt.plot(House_price_df_monthly.index, House_price_df_monthly['Price'])
plt.xlabel('Month')
plt.ylabel('Price')
plt.title('Price changes over the month')
plt.show()
```

From the graph, the prices of the houses generally decreases from June, 2016 to February, 2017. This maybe due to the low demand of houses by the Indians.

Let's move the target variable to the last column in order to ease data manipulation

```python
df = House_price_df.pop('Price')
House_price_df['Price'] = df
```

Now, we can identify the input and target columns

```python
House_price_df.columns
```

```python
input_cols, target_col = House_price_df.columns[:-1], House_price_df.columns[-1]
inputs_df, targets = House_price_df[input_cols].copy(), House_price_df[target_col].copy()
```

```python
numeric_cols = House_price_df[input_cols].select_dtypes(include=np.number).columns.tolist()
categorical_cols = House_price_df[input_cols].select_dtypes(include='object').columns.tolist()
```

```python
print(numeric_cols)
```

```python
print(categorical_cols )
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
imputer = SimpleImputer().fit(inputs_df[numeric_cols])
inputs_df[numeric_cols] = imputer.transform(inputs_df[numeric_cols])
scaler = MinMaxScaler().fit(inputs_df[numeric_cols])
inputs_df[numeric_cols] = scaler.transform(inputs_df[numeric_cols])
```

Since there are no missing values, one can ignore using imputer library.

```python
train_inputs, val_inputs, train_targets, val_targets = train_test_split(
    inputs_df[numeric_cols], targets, test_size=0.25, random_state=42)
```

# **Model**

## Ridge Regression Technique

```python
from sklearn.linear_model import Ridge
```

```python
# Create the model
# Initialize the model
alpha = 0.5
fit_intercept = True
model = Ridge(alpha=alpha, fit_intercept=fit_intercept)
```

```python
# Fit the model using inputs and targets
# Fit the model on the training data
model.fit(train_inputs, train_targets)
```

### Make Predictions and Evaluate Your Model

```python
from sklearn.metrics import mean_squared_error
```

```python
train_preds = model.predict(train_inputs)
```

```python
train_preds
```

```python
train_rmse = mean_squared_error(train_targets, train_preds, squared=False)
```

```python
print('The RMSE loss for the training set is $ {}.'.format(train_rmse))
```

```python
val_preds = model.predict(val_inputs)
```

```python
val_preds
```

```python
val_rmse = mean_squared_error(val_targets, val_preds, squared=False)
```

```python
print('The RMSE loss for the validation set is $ {}.'.format(val_rmse))
```

### Ridge Regression Feature Importance

Let's look at the weights assigned to different columns, to figure out which columns in the dataset are the most important.

```python
model.coef_
```

```python
model.intercept_
```

```python
weights = model.coef_
```

```python
weights_df = pd.DataFrame({
    'columns': train_inputs.columns,
    'weight': weights
}).sort_values('weight', ascending=False)
```

```python
weights_df.head(10)
```

Now, one can tell which columns have the greatest impact on the house prices.

```python
plt.title('Ridge Regression Feature Importance')
sns.barplot(data=weights_df.head(10), x='weight', y='columns');
```

## Random Forests

Random forests are machine learning techniques that work by averaging the results of several decision trees. The whole idea is that, each decision tree in the forest will generate different kinds of errors and upon averaging, many of these errors will cancel out. For this project, we'll use the RandomForestRegressor class from sklearn.ensemble.

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf1 = RandomForestRegressor(n_jobs = -1, random_state=42)
```

n_jobs permits the random forest to use multiple parallel workers to train decision trees, and random_state = 42 parameter ensures that we receive the same results for each execution.

Let's fit our model to the training data.

```python
%%time
rf1.fit(train_inputs, train_targets)
```

Now we generate predictions on the training and validation sets using the trained random forest, and compute the Root Mean Squared Error (RMSE) loss.

```python
rf1_train_preds = rf1.predict(train_inputs)
```

```python
rf1_train_rmse = mean_squared_error(train_targets, rf1_train_preds, squared=False)
```

```python
rf1_val_preds = rf1.predict(val_inputs)
```

```python
rf1_val_rmse = mean_squared_error(val_targets, rf1_val_preds, squared=False)
```

```python
print('Train RMSE: {}, Validation RMSE: {}'.format(rf1_train_rmse, rf1_val_rmse))
```

```python
rf1.score(train_inputs, train_targets), rf1.score(val_inputs, val_targets)
```

Here, we can see that the RMSE loss for our train data is 48062.023716204996, and the RMSE loss for our validation data is 140709.18596951824

```python
rf1_importance_df = pd.DataFrame({
    'feature': train_inputs.columns,
    'importance': rf1.feature_importances_
}).sort_values('importance', ascending=False)
```

```python
rf1_importance_df.head(10)
```

```python
sns.barplot(data=rf1_importance_df, x='importance', y='feature')
```

From the graph one can notice that the grade of the house, living area, latitudes, and longitudes are the most important input variables for this model.
